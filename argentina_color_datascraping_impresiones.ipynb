```python
import pandas as pd
from bs4 import BeautifulSoup
import smtplib
import requests
import time
import datetime
```

```python
# celulares

URL = "https://www.argentinacolor.com/impresion"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:103.0) Gecko/20100101 Firefox/103.0"}
page = requests.get(URL,headers=headers)

soup1 = BeautifulSoup(page.content, "html.parser")
soup2 = BeautifulSoup(soup1.prettify(),"html.parser")
```

```python
prices = soup2.find_all('span', attrs={'class':'price-tag-text-sr-only'})
names = soup2.find_all('h2', attrs={'class':'ui-search-item__title'})
print(prices)
print(names)


```

```python
# get price list,name list,date list
pricelist=[]
for span in prices:
    print(span.string)
    pricelist.append(span.string.strip())

namelist=[]
for h2 in names:
    print(h2.string)
    namelist.append(h2.string.strip())
    

today = datetime.date.today()

datelist=[]
for i in names:
    date = str(today)
    datelist.append(date)
    
```

```python
# remove values with "antes"
pricelistp = [x for x in pricelist if "Antes:" not in x]
print(pricelistp)    
```

```python
print(len(pricelistp))
print(len(namelist))
```

```python
joindata = list(zip(pricelistp,namelist,datelist))
print(joindata)
```

```python
# export data to csv

import csv
fields = ['price','product','date']
rows = joindata

with open ('ArgColorImpresiones.csv', 'w', newline='') as f:
    write = csv.writer(f)
    write.writerow(fields)
    write.writerows(rows)
```

```python
def check_price():
    # import soup

    URL = "https://www.argentinacolor.com/impresion"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:103.0) Gecko/20100101 Firefox/103.0"}
    page = requests.get(URL,headers=headers)

    soup1 = BeautifulSoup(page.content, "html.parser")
    soup2 = BeautifulSoup(soup1.prettify(),"html.parser")


    # filter soup
    prices = soup2.find_all('span', attrs={'class':'price-tag-text-sr-only'})
    names = soup2.find_all('h2', attrs={'class':'ui-search-item__title'})


    # get price list,name list,date list

    pricelist=[]
    for span in prices:
        print(span.string)
        pricelist.append(span.string.strip())

    namelist=[]
    for h2 in names:
        print(h2.string)
        namelist.append(h2.string.strip())
    

    today = datetime.date.today()

    datelist=[]
    for i in names:
        date = str(today)
        datelist.append(date)
    

    # remove values with "antes"
    pricelistp = [x for x in pricelist if "Antes:" not in x]
    print(pricelistp)


    # join data in a list
    joindata = list(zip(pricelistp,namelist,datelist))


    # export list to csv

    import csv
    fields = ['price','product','date']
    rows = joindata

    with open ('ArgColorImpresiones.csv', 'a+', newline='') as f:
        write = csv.writer(f)
        write.writerow(fields)
        write.writerows(rows)
```

```python
# create daily check (sleep would be 86400 for a daily check, set to 5 seconds to try it out)

while(True):
    check_price()
    time.sleep(5)
```
